```python
import numpy as np
import os
from matplotlib import pyplot as plt
from matplotlib.colors import rgb_to_hsv
from imageio import imread
```

```python
images = os.listdir('../image/')
```

```python
pink_hue = 328 / 360 * np.pi * 2
```

```python
len(images)
```

```python
images.index('C10100020003-8806198738242.png')
```

```python
result = dict()
for i in images:
    result[i] = ''
```

```python
for img in images[152:]:
    rgb = imread(f'../image/{img}')
    if rgb.shape[2] != 3:
        result[img] = 'Error'
        continue
    hsv = rgb_to_hsv(rgb)
    hue_list = []
    for i in range(500):
        for h, s, v in hsv[:,:,:][i]:
            if v >= 254 and s <= 2:
                continue
            hue = h * np.pi * 2
            cossim = np.cos(hue-pink_hue)
            hue_list.append(cossim)
    result[img] = np.mean(hue_list)
```
