```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/final_cossim_result.csv')
df.head()
```

```python
df.shape
```

```python
df = df[df['pink_cossim'] != 'Error']
```

```python
df = df[df['pink_cossim'] != 'Error2']
```

```python
df[df['blue_cossim'] == 'Error2']
```

```python
df = df[df['blue_cossim'] != 'Error2']
```

```python
df[df['blue_cossim'] == 'Error']
```

```python
df = df[df['blue_cossim'] != 'Error']
```

```python
df[['pink_cossim', 'blue_cossim']] = df[['pink_cossim', 'blue_cossim']].astype('float64')
```

```python
df.info()
```

```python
df_gender = df.groupby(df.성구분).mean()
df_gender
```

```python
df['pink_bin'] = pd.qcut(df['pink_cossim'], 4, labels=['never pinky','not pinky','pinky','the most pinky'])
df.head()
```

```python
df['blue_bin'] = pd.qcut(df['blue_cossim'], 4, labels=['never blue', 'not blue', 'bluish', 'the most bluish'])
df.head()
```
