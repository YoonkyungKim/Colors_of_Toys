```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from imageio import imread
import pandas as pd
```

```python
import cv2
import numpy as np
from pylab import figure, axes, pie, title, show
from PIL import Image
```

```python
df = pd.read_csv('data/final_cossim_result.csv')
```

```python
df.head()
```

```python
girl_paths = [df['파일명'].str.contains('C1010002')]
```

```python
boy_paths = [df['파일명'].str.contains('C1010003')]
```

```python
import glob
girl_paths = glob.glob("../image/C1010002*.png")
boy_paths = glob.glob("../image/C1010003*.png")
```

```python
len(girl_paths)
```

```python
len(boy_paths)
```

```python
dst_girl = mpimg.imread(girl_paths[0], 0)
numb = 1
for im in girl_paths[1:]:
    try:
        numb +=1
        alpha = (numb-1)*1/numb
        src = mpimg.imread(im, 0)
        dst_girl = cv2.addWeighted(dst_girl, alpha, src, 1-alpha, 0)
    except OSError:
        pass
    except:
        print('error')
        continue
```

```python
dst_boy = mpimg.imread(boy_paths[0], 0)
numb = 1
for im in boy_paths[1:]:
    try:
        numb +=1
        alpha = (numb-1)*1/numb
        src = mpimg.imread(im, 0)
        dst_boy = cv2.addWeighted(dst_boy, alpha, src, 1-alpha, 0)
    except OSError:
        pass
    except:
        print('error')
        continue
```
