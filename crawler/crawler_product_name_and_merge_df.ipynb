```python
import requests
import json
import pandas as pd
```

```python
df = pd.read_csv('../color analysis/data/cossim_result_with_IDs_eng.csv')
```

```python
df.head()
```

```python
product_ids = df['ProductID'].tolist()
category_ids = df['CategoryID'].tolist()
```

```python
category_id = category_ids[3]
product_id = product_ids[3]
```

```python
url = f'http://toysrus.lottemart.com/product/ProductDetail.do?CategoryID={category_id}&ProductCD={product_id}'
```

```python
texts = requests.get(url).text
```

```python
texts.split('<title>')[1].split('</title>')[0].split(' : 세계 최대 장난감ㆍ유아용품 전문점 토이저러스 온라인몰')[0]
```

```python
titles = list()
```

```python
for c, p in zip(category_ids[:10], product_ids[:10]):
    url = f'http://toysrus.lottemart.com/product/ProductDetail.do?CategoryID={c}&ProductCD={p}'
    texts = requests.get(url).text
    title = texts.split('<title>')[1].split('</title>')[0].split(' : 세계 최대 장난감ㆍ유아용품 전문점 토이저러스 온라인몰')[0]
    titles.append(title)
```

```python
titles
```

```python
for c, p in zip(category_ids[10:], product_ids[10:]):
    url = f'http://toysrus.lottemart.com/product/ProductDetail.do?CategoryID={c}&ProductCD={p}'
    texts = requests.get(url).text
    title = texts.split('<title>')[1].split('</title>')[0].split(' : 세계 최대 장난감ㆍ유아용품 전문점 토이저러스 온라인몰')[0]
    titles.append(title)
```

```python
len(titles)
```

```python
for i, t in enumerate(titles):
    if t == '미판매상품 안내페이지':
        print(i)
```

```python
df['상품명'] = titles
```

```python
df = df[['성구분', '카테고리명', '상품명', 'CategoryID', 'ProductID', 'cossim']]
```

```python
df.head()
```

```python
df = df.rename(columns={'cossim' : 'pink_cossim'})
```

```python
df2 = pd.read_csv('../color analysis/data/cossim_result_blue_with_IDs_eng.csv')
```

```python
df2.head()
```

```python
df = df.drop(['blue_cossim'], axis=1)
```

```python
df.head()
```

```python
df = df.merge(df2)
df.head()
```

```python
df = df.rename(columns={'cossim' : 'blue_cossim'})
df.head()
```

```python
df = df[['성구분', '카테고리명', '상품명', 'CategoryID', 'ProductID', 'pink_cossim', 'blue_cossim', '파일명']]
df.head()
```
