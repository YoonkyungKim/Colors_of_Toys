```python
categs = categories['카테고리ID'].tolist()
total = categories['총 개수'].tolist()
num_category = dict()
for c, t in zip(categs, total):
    num_category[c] = t
num_category
```

```python
driver = webdriver.Chrome('C:/Users/rlath/Downloads/chromedriver')

for ID, num in num_category.items():
    for page_num in range(num//90+1):
        temp = pd.DataFrame()
        url = f'http://toysrus.lottemart.com/category/categoryList.do?currentPage={page_num+1}&CategoryID={ID}&rowPerPage=90'
        driver.get(url)
        time.sleep(1)

        html = driver.page_source
        bs = BeautifulSoup(html, 'html5lib')
        toys = bs.select('h3.goods-name')
        
        toys_name = [toy.text.replace('\n', '').replace('\t', '') for toy in toys]
        temp['제품명'] = toys_name
        temp.to_csv(f'{ID}_{page_num}.csv', index=False)
```
