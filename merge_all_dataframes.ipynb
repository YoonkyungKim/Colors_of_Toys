```python
import pandas as pd
```

```python
boys = pd.read_csv('data/colors_of_toys_category_boy.csv')
girls = pd.read_csv('data/colors_of_toys_category_girl.csv')
```

```python
children = boys.append(girls)
children
```

```python
categories = children['카테고리ID'].tolist()
categories
```

```python
boy_01_01 = pd.read_csv('data/before_merge/C10100020001_0.csv')
```

```python
len(categories), len([5,24,5,6,4,8,3,9,7])
```

```python
indexes = [4,8,3,9,7,5,24,5,6]
```

```python
category_dict = dict()
for ID, num in zip(categories, indexes):
    category_dict[ID] = num
```

```python
category_dict
```

```python
for ID in categories:
    for i in range(category_dict[ID]):
        df = pd.read_csv(f'data/before_merge/{ID}_{i}.csv')
        for index in df.index:
            df['카테고리ID'] = ID
            df.to_csv(f'data/temp_dfs/{ID}_{i}.csv', index=False)
```

```python
result = pd.DataFrame(columns='카테고리ID,제품명'.split(','))
for ID in categories:
    for i in range(category_dict[ID]):
        df = pd.read_csv(f'data/temp_dfs/{ID}_{i}.csv')
#         df = df[['카테고리ID', '제품명']]
        result = result.append(df)
result.shape
```
